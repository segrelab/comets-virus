```python
import pandas as pd
pd.set_option('display.max_rows', None)

import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
#Infinite Glucose 1000000000000.0

allData = pd.read_csv('ResourcePartitionB|static|uptake=1|vmin=1|Glc=500.csv')

allData["Host Lifespan"] /= 100
```

```python
plt.rcParams['font.size'] = 20

fig, ax1 = plt.subplots(figsize= (13, 7))

ax2 = ax1.twinx()
ax3 = ax1.twinx()

ax3.spines['right'].set_position(('axes', 1.2))
ax3.set_frame_on(True)
ax3.patch.set_visible(False)

ax1.set_ylabel('Final Virus Biomass (mmol)', color= 'red')
ax2.set_ylabel('Virus Production\nFlux (g Virus/g Host/hr)', color= 'blue')
ax3.set_ylabel('Host Lifespan (hrs)', color= 'green')

ax1.tick_params(axis='y', colors='red')
ax2.tick_params(axis='y', colors='blue')
ax3.tick_params(axis='y', colors='green')

ax1.set_xlabel("Virus ATP Partition Ratio")

sns.lineplot(data= allData, x= "Virus Partition Ratio", y= "Virus Mass", color= 'red', ax= ax1, linewidth = 3)
sns.lineplot(data= allData, x= "Virus Partition Ratio", y= "Virus Rate", color= 'blue', ax= ax2, linewidth = 3)
sns.lineplot(data= allData, x= "Virus Partition Ratio", y= "Host Lifespan", color= 'green', ax= ax3, linewidth = 3)

ax1.set_ylim(0, 7.5e-7)
ax2.set_ylim(0, 0.025)
ax3.set_ylim(0, 10)

plt.show()
```
