```python
import pandas as pd
pd.set_option('display.max_rows', None)

import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
#Infinite Glucose 1000000000000.0

allData = pd.DataFrame()

for vmin in [0.75, 1, 1.25, 1.5, 1.75]:
    d = pd.read_csv('ResourcePartitionB|static|uptake=1|vmin=' + str(vmin) + '|Glc=500.csv')
    allData = pd.concat([allData, d])
    
allData.reset_index(drop=True, inplace=True)

allData["Host Lifespan"] /= 100
```

```python
plt.rcParams['font.size'] = 11
fig, axs = plt.subplots(ncols= 2, nrows= 1, figsize= (8, 3), dpi= 300)

sns.lineplot(data= allData, x= "Virus Partition Ratio", y= "Virus Mass", hue= "ATP Threshold", palette= 'flare', ax= axs[0], linewidth = 2)
#plt.legend(bbox_to_anchor=(1, 1), title= "ATP Threshold")
#plt.legend(bbox_to_anchor=(1.42, 1))
axs[0].set_xlabel('Virus ATP Partition Ratio')
axs[0].set_ylabel('Final Virus Biomass (mmol)')
axs[0].set_title('A', loc= 'left', pad= 30, fontsize= 15)
axs[0].get_legend().set_visible(False)

sns.lineplot(data= allData, x= "Virus Partition Ratio", y= "Host Lifespan", hue= "ATP Threshold", palette= 'flare', ax= axs[1], linewidth = 2)
plt.legend(['0, 0.75', '1.0', '1.25', '1.5', '1.75'], bbox_to_anchor=(1, 1), title= "ATP Requirement", )
axs[1].set_xlabel('Virus ATP Partition Ratio')
axs[1].set_ylabel('Host Lifespan (hrs)')
axs[1].set_title('B', loc= 'left', pad= 30, fontsize= 15)
```
